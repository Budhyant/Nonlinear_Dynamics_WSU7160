$10\ddot{x} + 1\dot{x} + 100x = 0 \quad , \quad x(0) = 5 \quad , \quad \dot{x}(0) = 2$

$\ddot{x} + 0.1 \dot{x} + 10x = 0 \rightarrow \zeta = \dfrac{0.1}{2\sqrt(10)} \quad , \quad \omega_n = \sqrt{10}$
$\quad , \quad \omega_d = \omega_n \sqrt{1 - \zeta^2} \quad \longrightarrow \quad$

$x(t) = A e^{-\zeta \omega_n t} \sin(\omega_d t + \phi) \quad , \quad$
$A = \sqrt{\dfrac{(v_0 + \zeta \omega_n x_0)^2 + (x_0 \omega_d)^2}{\omega_d^2}} \quad , \quad$
$\phi = \tan^{-1} \dfrac{x_0 \omega_d}{v_0 + \zeta \omega_n x_0} \quad \longrightarrow$
$x(t) = 5.05e^{-0.05t} \sin(3.16t + \phi)$

Damping ration ($\zeta$) = 0.005

Time constant ($\tau$) = 20.0

Settling Time($T_s$) = $\dfrac{3.9}{\zeta \omega_N}$ = 78

Natural Frequency ($\omega$) = 10.0

```python
def q1B():
    # time definition
    t0 = 0.0
    tEnd = 120.0
    nt = 1000

    # response definition
    x0 = 5
    v0 = 2
    zeta = 0.1 / (2 * pl.sqrt(10))
    omegaN = pl.sqrt(10)
    omegaD = omegaN * pl.sqrt(1 - np.power(zeta, 2))
    A = np.sqrt( (np.power(v0 + zeta * omegaN * x0, 2) + \
                  np.power(x0 * omegaD, 2)) / \
                 np.power(omegaD, 2) )
    phi = np.arctan(x0 * omegaD / (v0 + zeta * omegaN * x0))

    # Define the response function
    def x(t):
        X = A * np.exp(-zeta * omegaN * t) * \
            np.sin(omegaD * t + phi)
        return X
    # END of function definition

    t = np.linspace(t0, tEnd, nt, endpoint = True)
    x = x(t)

    plt.figure(figsize=(20,10))
    plt.plot(t, x, linewidth = 2.5, color = 'k')
    pl.xlabel('$\mathbf{t}$')
    pl.ylabel('$\mathbf{X(t)}$')
    pl.title('$\mathbf{X(t)} = \mathbf{5.05e^{-0.05t} \sin(3.16t + \phi)}$')
    pl.show()
    
q1B()
```

Assume solution has a form like: $x = A e^{\lambda t}$

$(10 \lambda^2 + 2 \lambda + 100) A e^{\lambda t} = 0 \quad$ Homogeneous solution

$10 \lambda ^ 2 + 2 \lambda + 100 = 0 \quad \rightarrow \quad \lambda_1 = \dfrac{-1 + i\sqrt{999}}{10} \quad , \quad \lambda_2 = \dfrac{-1 - i\sqrt{999}}{10}$

Particular solution can be written as:

$x(t) = Ae^{\lambda_1 t} + Be^{\lambda_1 t} + 2$

After applyting the boundary conditions we get:

$A = -1 + 0.03i \quad , \quad B = -1 - 0.03i$

Damping ration ($\zeta$) = 0.1

Time constant ($\tau$) = 0.1

Settling Time($T_s$) = $\dfrac{3.9}{\zeta \omega_N}$ = 3.9

Natural Frequency ($\omega$) = 10

```python
def q1D():
    lambda1 = (-1 + 1j * np.sqrt(999)) / 10
    lambda2 = (-1 - 1j * np.sqrt(999)) / 10

    A = np.array([[1, 1], [lambda1, lambda2]])
    b = np.array([-2, 0])
    X = np.linalg.solve(A, b)

    def x(t):
        return X[0] * np.exp(lambda1 * t) + \
               X[1] * np.exp(lambda2 * t) + \
               2

    t0 = 0.0
    tEnd = 60.0
    nt = 1000
    t = np.linspace(t0, tEnd, nt, endpoint = True)

    plt.figure(figsize = (20,10))
    plt.plot(t, np.real(x(t)), linewidth = 2.5, color = 'k')
    plt.xlabel('$\mathbf{t}$')
    plt.ylabel('$\mathbf{X(t)}$')
    plt.title('$x(t) = Ae^{\lambda_1 t} + Be^{\lambda_1 t} + 2$')
    plt.show()
    plt.close()
    
q1D()
```

## 3C)

We used backward finite difference to solve this equation. The discretized form can be written as:

$\dot{x} + x - 0.5x^3 = 0 \quad \rightarrow \quad$
$\dfrac{x_{n+1} - x_n}{\Delta t} + x_n - 0.5x_n^3 = 0$
$\quad \rightarrow \quad$

$x_{n+1} = (0.5x_n^3 - x_n) \Delta t + x_n$

```python
t0 = 0.0
tEnd = 4.0
dt = 0.00001

nt = (tEnd - t0) / dt
t = np.linspace(t0, tEnd, nt + 1, endpoint=True)


def calcX(t, x0 = 0):
    x = np.zeros(len(t))
    x[0] = x0
    for n in range(0,len(x)-1):
        x[n+1] = (0.5 * np.power(x[n],3) - x[n]) * dt + x[n]
    fig = plt.figure(figsize=(20,10))
    plt.plot(t, x, 'k', linewidth=2.0)
    plt.xlabel('$t$')
    plt.ylabel('$X(t)$')
    plt.title('Time history of $X$')

    plt.figure(figsize=(20,10))
    plt.plot(x, 0.5 * np.power(x, 3) - x, 'k-',
             x[0], 0.5 * np.power(x[0], 3) - x[0], 'ro',
             ms=20, mew=3,
             linewidth=2.0)
    plt.xlabel('$X(t)$')
    plt.ylabel('$Y(t)$')
    plt.title('Phase plot')
    plt.show()
    return fig
```

We looked at the responce of the system to different initial conditions. This system has three equilibrium locations, $-\sqrt{2}$, $0$, and $\sqrt{2}$. We plot the results for initial locations near these locations. The initla location is shown by a red circle in the phase plot.

```python
x = calcX(t, x0 = 0) # x0 is the initial condition
```

```python
x = calcX(t, x0 = 1) # x0 is the initial condition
```

```python
x = calcX(t, x0 = np.sqrt(2)) # x0 is the initial condition
```

```python
x = calcX(t, x0 = np.sqrt(2) + 0.0001) # x0 is the initial condition
```

As shown in above plots, the system is stable for initial conditions between $-\sqrt{2}$ and $\sqrt{2}$. The responce becomes unstable for points ourside this region.

```python
def calcX(mu=0.1, t0=0.0, tEnd=10.0, dt=0.001, x0=1.0, v0=0.0, plotInfo='all'):
    nt = int((tEnd - t0) / dt)
    X = np.zeros(shape=(2, nt + 1))
    t = np.linspace(t0, tEnd, nt + 1, endpoint=True)
    X[0, 0] = x0
    X[1, 0] = v0
    # First order backward Euler method
    for n in range(0, nt):
        A = np.array([[0.0, 1.0], [-1.0, -mu * (np.power(X[0, n], 2) - 1.0)]])
        X[:, n+1] = np.dot(A * dt + np.eye(2, 2), X[:, n])

    # Plotting
    if (plotInfo == 'displacement') | (plotInfo == 'all'):
        plt.figure(figsize=(20,10))
        plt.plot(t, X[0, :], 'k-', lw=2.0)
        plt.title('Displacement time history')
        plt.xlabel('t')
        plt.ylabel('X(t)')

    if (plotInfo == 'velocity') | (plotInfo == 'all'):
        plt.figure(figsize=(20,10))
        plt.plot(t, X[1, :], 'k-', lw=2.0)
        plt.title('Velocity time history')
        plt.xlabel('t')
        plt.ylabel('V(t)')

    if (plotInfo == 'phase') | (plotInfo == 'all'):
        plt.figure(figsize=(20,10))
        plt.plot(X[0, :], X[1, :], 'k-',
                 X[0,-1], X[1,-1], 'r+', lw=2.0, ms=20.0, mew=5.0)
        plt.title('Phase diagram')
        plt.xlabel('X(t)')
        plt.ylabel('V(t)')
    
    plt.show()
```

```python
calcX(mu=-0.5, t0=0.0, tEnd=200.0, dt=0.001, x0=0.0, v0=1.0, plotInfo='all')
```
