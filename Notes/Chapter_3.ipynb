## Floquet Theory

### Autonomous Systems

Consider

$$\dot{\mathbf{x}} = \mathbf{F}(\mathbf{x};\mathbf{M})$$

Define the periodic solution when $\mathbf{M} = \mathbf{M}_0$ to be $\mathbf{X}_0(t)$ with a minimal period $T$ (lowest value of $t$ for which the cycle repeats). 

A disturbance $\mathbf{y}(t)$ is imposed on $\mathbf{X}_0$ so that

$$\mathbf{x}(t) = \mathbf{X}_0(t) + \mathbf{y}(t)$$

Substituting into the governing equation and expanding in a Taylor series about $\mathbf{X}_0(t) $

$$\dot{\mathbf{y}}(t) = D_\mathbf{X}\mathbf{F}(\mathbf{x};\mathbf{M}_0)\mathbf{y}(t)+O(||\mathbf{y}||^2)$$

or
$$\dot{\mathbf{y}}(t) \approx A(t;\mathbf{M}_0)\mathbf{y}(t)$$

We now have a state equation in the behavior in the vicinity of $\mathbf{X}_0$ and can work towards determining the stability of the solution $\mathbf{X}_0$.

- The matrix $A$ is periodic in time (repeating each cycle) and has a period of $T$. 
- However, $T$ may not be the *minimal* period. 
- Example: for the Duffing Oscillator, the odd nonlinearity causes a period of $\frac{1}{2}T$ to exist.

Floquet theory deals with such systems as $\dot{\mathbf{y}}(t) \approx A(t;\mathbf{M}_0)\mathbf{y}(t)$. 

The $n$-dimensional linear system has $n$ linearly independent solutions $\mathbf{y}_i(t)$ referred to as the **fundamental set of solutions**:

$$Y(t) = \left[ \mathbf{y}_1(t), \mathbf{y}_2(t), \mathbf{y}_3(t),\cdots, \mathbf{y}_n(t) \right]$$

These must (of course) satisfy the equation so that

$$\dot{Y} = A(t;\mathbf{M}_0) Y$$

We would like to changed the variable $t$ to $\tau = t+T$ giving

$$\frac{dY}{d\tau} = A(\tau-T;\mathbf{M}_0)Y=A(\tau;\mathbf{M}_0)Y$$

because $A(\tau-T;\mathbf{M}_0)=A(\tau;\mathbf{M}_0)$

Also, since
$$Y(t) = \left[ \mathbf{y}_1(t), \mathbf{y}_2(t), \mathbf{y}_3(t),\cdots, \mathbf{y}_n(t) \right]$$
is a fundamental solution, then so is 

$$Y(t+T) = \left[ \mathbf{y}_1(t+T), \mathbf{y}_2(t+T), \mathbf{y}_3(t+T),\cdots, \mathbf{y}_n(t+T) \right]$$
 
a solution. 

Further, since $\mathbf{y}_i(t)$ are $n$ linearly independent solutions, then each $\mathbf{y}_i(t+T)$ must be a linear combination of the  $\mathbf{y}_i(t)$. 

So, we could then write that

$$Y(t+T) = Y(t)\Phi$$

where $\Phi$ is an $n\times n$ constant matrix that depends on the chosen fundamental solutions $\mathbf{y}_i(t)$. 

This could be thought of as a map (recall maps?) that transforms an initial vector at $t=0$ to another vector at $t=T$. 

Let's specify an initial condition 

$$Y(0) = I$$
where $I$ is the identity matrix

In doing so, we can write that

$$\Phi = Y(T)$$

The matrix $\Phi$ defined this way is called the **monodromy matrix**. 

Introducing the coordinate transformation

$$Y(t) = V(t) P^{-1}$$

where $P$ is a nonsingular constant $n\times n$ matrix, we can write

$$V(t+T)=V(t)J$$
where
$$J = P^{-1}\Phi P$$

we choose $J$ to have the simplest possible form, which depends on the eigenvalues and eigenvectors of $\Phi$. 

When the eigenvalues of $\Phi$ are distinct, one can chose $P$ such that the columns are the right eigenvectors of $\Phi$ corresponding to eigenvalues $\rho$ such that

$$\Phi \mathbf{p}_m=\rho_m\mathbf{p}_m$$

thus 

$$P = \left[\mathbf{p}_1  \mathbf{p}_2 \mathbf{p}_3\cdots \mathbf{p}_n \right]$$

With this choice

\begin{split}
J &= P^{-1}\Phi P = P^{-1}\Phi \left[\mathbf{p}_1  \mathbf{p}_2 \mathbf{p}_3\cdots \mathbf{p}_n \right]\\
&=P^{-1} \left[\Phi\mathbf{p}_1  \Phi\mathbf{p}_2 \Phi\mathbf{p}_3\cdots \Phi\mathbf{p}_n \right]\\
&=P^{-1} \left[\rho_1\mathbf{p}_1  \rho_2\mathbf{p}_2 \rho_3\mathbf{p}_3\cdots \rho_n\mathbf{p}_n \right]\\
&=P^{-1}P D = D
\end{split}

where

$$D = \begin{bmatrix}
\rho_1&0&\cdots&0\\
0&\rho_2&\cdots&0\\
\vdots&\vdots&\ddots&\vdots\\
0&0&\cdots&\rho_n\\
\end{bmatrix}$$

The eigenvectors and eigenvalues need not be real!

The eigenvalues $\rho_m$ of $\Phi$ are called **Floquet** or **characteristic multiplier** and are unique for a given $A$. 

Each $\rho_m$ provides a measure of the **local orbital divergence** or **convergence** along a particular direction over one period of the closed orbit. 

When we have distinct Floquet multipliers, we can write

$$\mathbf{v}_m(t+T) = \rho_m \mathbf{v}_m(t)$$


which is a first order differential equation in $\mathbf{v}_m(t)$ in discrete form. 

Further, we can repeatedly apply this equation to result in

$$\mathbf{v}_m(t+N T) = \rho_m^N \mathbf{v}_m(t)$$

So:

$$\mathbf{v}_m(t)\to 0 \, \text{if}\,|\rho_m|<1$$
and
$$\mathbf{v}_m(t)\to \infty \, \text{if}\,|\rho_m|>1$$

When $\rho_m=1$, $\mathbf{v}_m(t)$ has a periodicity of $T$, and if $\rho_m=-1$, $\mathbf{v}_m(t)$ has a periodicity of $2T$ (according to Nayfeh Balachandran)

**One of the Floquet multipliers associated with a periodic solution $\mathbf{X}_0(t)$ of an autonomous system of equations is always $1$.**

Differentiate the original governing equation with respect to time:

Using the chain rule we get
$$\ddot{\mathbf{x}} = D_\mathbf{x}\mathbf{F}(\mathbf{x};\mathbf{M})\dot{\mathbf{x}}$$

if $\mathbf{x}(t)$ is a solution of the original equation, then $\dot{\mathbf{x}}$ must be a solution to this expression, which you should recall was the same equation for $\mathbf{y}$. 

Thus $\dot{\mathbf{X}}_0(t)$ is therefore also a solution. 

Since 

$$\mathbf{X}_0(t) = \mathbf{X}_0(t+T)$$

then 

$$\dot{\mathbf{X}}_0(t) = \dot{\mathbf{X}}_0(t+T)$$

hence
$$\dot{\mathbf{X}}_0(0) = \dot{\mathbf{X}}_0(T)$$

Note that we earlier derived that

$$\dot{\mathbf{y}} = D_\mathbf{x}\mathbf{F}(\mathbf{y};\mathbf{M}){\mathbf{y}}$$

which we just wrote above with $\mathbf{x}$ in place of $\mathbf{y}$, so, $\dot{\mathbf{X}}_0(t)$
is a solution of this equation as well, and must also be a linear combination of $\mathbf{y}_i(t)$. 

So

$$\dot{\mathbf{X}}_0(t) = Y(t)\alpha$$

where $\alpha$ is a constant vector. 

evaluating at $t=0$ and $t=T$

$$\dot{\mathbf{X}}_0(0) = Y(0)\alpha$$

and
$$\dot{\mathbf{X}}_0(T) = Y(T)\alpha$$

Since $\dot{\mathbf{X}}_0(0) = \dot{\mathbf{X}}_0(T) $,

$$Y(T)\alpha = Y(0)\alpha$$

Recalling that $Y(0)=I$ and $Y(T)=\Phi$, 

$$\Phi\alpha = \alpha$$

Therefore, as the eigenvalue problem is defined at

$$\Phi\alpha = \lambda\alpha$$
 with an eigenvalue $\lambda$, at least one eigenvalue of $\Phi$ must be equal to $1$ with the corresponding eigenvector
 
 $$\alpha = \dot{\mathbf{X}}_0(0) = \mathbf{F}(\mathbf{X}_0;\mathbf{M}_0)$$


This corresponds to a disturbance along the trajectory of the system. Such a disturbance neither expands nor decays. 

#### Hyperbolic periodic solutions


- A periodic solution such as $\mathbf{X}_0$ is called a **hyperbolic periodic solution** if only one Floquet multiplier is located on the unit circle in the complex plain. 
- It may be either **stable** or **unstable**. 
- It is asymptotically stable if their are no Floquet multipliers outside the unit circle in the complex plane. 
  - This means all adjacent orbits are attracted to this orbit. 
  - This is a **periodic attractor** or **stable limit cycle**. 
- It is unstable if one or more Floquet multipliers are outside the unit circle in the complex plane. 
  - This is a **periodic repellor** or **unstable limit cycle**. 
  
We can have an **unstable limit cycle of the saddle type**. 

#### Nonhyperbolic periodic solutions
If two or more Floquet multipliers are located on the unit circle, the periodic solution is a **nonhyperbolic periodic solution**. 

- If any of the other eigenvalues are outside the unit circle, then the orbit is unstable. 
- If all of the others are inside the unit circle, then nonlinear analysis is necessary to determine stability of the solution. 

Recall

$$\mathbf{v}_m(t+T) = \rho_m\mathbf{v}_m(t)$$

Premultiplying by $e^{-\gamma_m(t+T)}$

$$e^{-\gamma_m(t+T)}\mathbf{v}_m(t+T) = \rho_m e^{-\gamma_m(t+T)}\mathbf{v}_m(t)$$

Define $\gamma_m$ so that

$$\rho_m=e^{\gamma_mT}\;\text{or}\;\gamma_m = \frac{1}{T}\log_n\rho_m$$
Substituting in the the prior equation results in 

$$e^{-\gamma_m(t+T)}\mathbf{v}_m(t+T) = e^{-\gamma_m (t) }\mathbf{v}_m(t)$$

As a result, we can define $\mathbf{\phi}_m(t) = e^{-\gamma_m(t)}\mathbf{v}_m(t)$ which is a periodic vector with a period of $T$ (**$\mathbf{\phi}_m(t)$ is not related to $\Phi$!**)  so that

$$\mathbf{\phi}_m(t+T)= \mathbf{\phi}_m(t)$$

$\mathbf{v}_m(t)$ can be expressed in the **normal** or **Floquet form**

$$\mathbf{v}_m(t) = e^{\gamma_m t}\mathbf{\phi}_m(t)$$

The values $\gamma_m$ are called **characteristic exponents** and are unique within an integer multiple of $2\pi i/T$. 

If the real part of $\gamma_m$ is negative, then 

$$\mathbf{v}_m(t)\to 0\;\text{as}\; t\to\infty$$

If the real part  of $\gamma_m$ is positive, then 

$$\mathbf{v}_m(t)\to \infty\;\text{as}\; t\to \infty$$

The Floquet form can be used to solve 

$$\dot{\mathbf{y}}(t) = D_\mathbf{X}\mathbf{F}(\mathbf{x};\mathbf{M}_0)\mathbf{y}(t)+O(||\mathbf{y}||^2) \approx A(t;\mathbf{M}_0)\mathbf{y}$$
by seeking a solution of the form

$$\mathbf{y}(t) = e^{\gamma_ t}\mathbf{\phi}(t)$$

where $\mathbf{\phi}(t+T) = \mathbf{\phi}(t)$. Substituting $\mathbf{y}(t) = e^{\gamma_m t}\mathbf{\phi}(t)$ into the differential equation yields

$$\dot{\mathbf{\phi}}=[A(t;\mathbf{M}_0)-\gamma I]\mathbf{\phi}$$

after noting that $\dot{\mathbf{y}} = \gamma e^{\gamma_ t}\mathbf{\phi}(t)+e^{\gamma_ t}\dot{\mathbf{\phi}}(t)$

Expanding $\mathbf{\phi}(t)$ as a Fourier series

$$\mathbf{\phi}(t) =\sum_{k=-\infty}^{k=\infty}\mathbf{a}_k e^{\frac{2i\pi k t}{T}}$$

Substituting and equating harmonics we obtain an infinite-dimension eigenvalue for eigenvalues $\gamma$ and eigenvectors $\mathbf{a}_k$. The determinant of the coefficient matrix is called **Hill's determinant**.

When the coefficients of the time varying terms of $A(t;\mathbf{M}_0)$ are small, one can use perturbation methods to determine an approximate solution instead. 

#### Example 3.10

Consider the solution of 

$$\dot{x} = \mu x-\omega y+(\alpha x-\beta y)(x^2 + y^2)$$
and
$$\dot{y} = \omega x+ \mu y+(\beta x+\alpha y)(x^2 + y^2)$$

given as

$$X_0(t) = \left(-\frac{\mu}{\alpha}\right)^{\frac{1}{2}}\cos\left[\left(\omega-\frac{\beta\mu}{\alpha}\right)t+  \theta_0\right]$$

and
$$Y_0(t) = \left(-\frac{\mu}{\alpha}\right)^{\frac{1}{2}}\sin\left[\left(\omega-\frac{\beta\mu}{\alpha}\right)t+  \theta_0\right]$$

If we substitute

$$x(t) = X_0(t)+\zeta_1(t)$$  and
$$y(t) = Y_0(t)+\zeta_2(t)$$  

and linearize in $\boldsymbol{\zeta}$ we get

$$\dot{\boldsymbol{\zeta}}=A(\theta)\boldsymbol{\zeta}$$

where $\boldsymbol{\zeta}$ is a two dimensional vector of $\zeta_i(t)$,

$$\theta = \left(\omega-\frac{\beta\mu}{\alpha}\right)t+\theta_0$$

and

\begin{split}A &=\\&\left[\begin{matrix}- \mu \cos{\left (2 \theta \right )} - \mu + \frac{\beta \mu}{\alpha} \sin{\left (2 \theta \right )} & - \mu \sin{\left (2 \theta \right )} - \omega - \frac{\beta \mu}{\alpha} \cos{\left (2 \theta \right )} + \frac{2 \beta}{\alpha} \mu\\- \mu \sin{\left (2 \theta \right )} + \omega - \frac{\beta \mu}{\alpha} \cos{\left (2 \theta \right )} - \frac{2 \beta}{\alpha} \mu & \mu \cos{\left (2 \theta \right )} - \mu - \frac{\beta \mu}{\alpha} \sin{\left (2 \theta \right )}\end{matrix}\right]
\end{split}

The components of $A$ are periodic in $T = 2\pi/(\omega-\beta\mu/\alpha)$, the period of the limit cycle. 

However, the minimal period is $T/2$. 

Replacing the independent vatiable $t$ by $\theta$ and noting that $\dot{\theta}=\omega-\frac{\beta\mu}{\alpha}$ we can rewrite the state equation as

$$\frac{d\boldsymbol{\zeta}}{d\theta} = \left(\omega-\frac{\beta\mu}{\alpha}\right)^{-1}A(\theta)\boldsymbol{\zeta}$$

To determine stability we now need only determine the stability of this equation as $\theta\to\infty$.

To do this, we calculate two solutions for $\boldsymbol{\zeta}$ by integrating from $\theta=0$. 

Consider the initial conditions

$$\boldsymbol{\zeta}^{(1)}(0) = \begin{bmatrix}1\\0\end{bmatrix}$$

and
$$\boldsymbol{\zeta}^{(2)}(0) = \begin{bmatrix}0\\1\end{bmatrix}$$

The monodromy matrix is then

$$\Phi = \left[\boldsymbol{\zeta}^{(1)}(2\pi)\;\;\boldsymbol{\zeta}^{(2)}(2\pi)\right]$$

The check that one of the eigenvalues of $\Phi$ is $1$ assures that we are doing this right. 

If the magnitude of the other eigenvalue is less than 1, then the periodic solution is stable. Otherwise... it's not. 

For $\omega = 1.0,\, \beta=1.0,\,\mu=1.0$ and $\alpha = -1.0$

$$\Phi = \begin{bmatrix}
0.00187&0.00085\\
0.99813&1.00000
\end{bmatrix}
$$
with eigenvalues of $1.00085$ and $0.00102$. Hence, the periodic solution is stable. 

When $\omega = 1.0,\, \beta=1.0,\,\mu=-1.0$ and $\alpha = 1.0$

$$\Phi = \begin{bmatrix}
535.49200&0.00000\\
534.49200&1.00000
\end{bmatrix}
$$

which has eigenvalues of $1.00000$ and $535.49200$. Hence, the periodic solution is instable for those parameters.

## Poincaré maps

To generate a Poincaré map, we first have to define a *Poincaré section*. 

A Poincaré section is a *hypersurface* in the state space that is transverse to the flow. Transverse just means that the state flow cuts through it. 

In an *n-dimensional* space, a hypersurface is a surface whose dimensions in less than *n*. For instance, in 3-d space, consider a wobbly orbit around a planet (neglecting how this might happen). The surface would be a plane that the orbits cut through, once per orbit. 

For transverse, we mathematically mean

$$\mathbf{n}(\mathbf{x})\cdot\mathbf{F}(\mathbf{x})\neq 0$$ 
or $$\mathbf{n}^{T}\!(\mathbf{x})\mathbf{F}(\mathbf{x})\neq 0$$ 

where $\mathbf{n}(\mathbf{x})$ is a vector normal to the section located at $(\mathbf{x}, \mathbf{F}(\mathbf{x}))$ is the vector field describing the flow, and the dot represents the dot product. 

#### Example 3.13

Consider constructing the Poincaré section for orbits of the system defined by 
the generic system 

$$\begin{bmatrix}
\dot{x}\\\dot{y}\\\dot{z}
\end{bmatrix}=\begin{bmatrix}F_1(x,y,z)\\F_2(x,y,z)\\F_3(x,y,z)
\end{bmatrix}
$$

We can define a section by $y=y_0$ where $y_0$ is a value that we choose. We must also  ensure that $\mathbf{n}(\mathbf{x})\cdot\mathbf{F}(\mathbf{x})\neq 0$. For this plane, 
<img src="Figure3.3.1.png">

$$\mathbf{n} = \begin{bmatrix}0\\1\\0\end{bmatrix}$$

and thus we have to ensure that

$$\begin{bmatrix}0\\1\\0\end{bmatrix}\cdot
\begin{bmatrix}F_1(x,y_0,z)\\F_2(x,y_0,z)\\F_3(x,y_0,z)
\end{bmatrix} = 
F_2(x,y_0,z)
\neq 0$$

The figure shows three intersections. If we construct the figure such that $F_2(y_0)>0$, we have a **one-sided section**, where without this requirement we have a **two-sided section**. 

The time between two successive intersections is not necessarily at evenly spaced periods. 

If we track the behavior of the points at each increment, we have a map, just as we did earlier in the semester. 

$$\{\mathbf{X}_1, \mathbf{X}_2, \mathbf{X}_3,\cdots\}$$

We could then identify a mapping function

$$\mathbf{X}_{m+1}=\mathbf{P}(\mathbf{X}_m)$$
called a *Poincaré map*.

Because (and we know this because the system is deterministic) that the relationship between $\mathbf{X}_{m+1}$ and $\mathbf{X}_{m}$ is unique and invertible, a Poincaré map is an invertible map. So, we also have

$$\mathbf{X}_{m}=\mathbf{P}^{-1}\mathbf{X}_{m+1}$$

Further, the map shares the properties of the flow described by the associated continuous-time system. So, the map is also an **orientation preserving diffeomorphism**.

Orientation preserving means that the determinant of $D\mathbf{P}$ is always positive.

If the flow is dissipative, the determinant of $D\mathbf{P}$ has a magnitude of less than 1. 

Poincaré maps can be very useful when the behavior is peculiar, as can happen in subharmonic resonances for example.

The figure on the right shows what a Poincaré map can look like in steady state when a subharmonic response exists along with a harmonic response to a harmonic excitation. 

### Nonautonomous Systems

In a nonautonomous system (meaning $\mathbf{F}$ is explicitely a function of time, for instance a forced excitation) we typically know the period of the response. 

#### Example 3.14

Consider a case where 

$$\begin{bmatrix}
\dot{x}_1\\\dot{x}_2
\end{bmatrix}
= \begin{bmatrix}
F_1(x_1,x_2)\\F_2(x_1,x_2,t)
\end{bmatrix}$$

Note that $F_2$ is also a function of time $t$. 

We can modify the governing equations and make them autonomous as before resulting in
$$\begin{bmatrix}
\dot{x}_1\\\dot{x}_2\\\dot{\theta}
\end{bmatrix}
= \begin{bmatrix}
F_1(x_1,x_2)\\F_2(x_1,x_2,\theta)\\\frac{2\pi}{T}
\end{bmatrix}$$
The solution of the last equation is of course $\theta = \frac{2\pi t}{T}+\theta_0$. We can then simply define the Poincaré section to be taken at times when $\frac{2\pi t}{T}=2\pi n$ where $n = 0, 1, 2, 3\ldots$.

#### Example 3.17
Consider the equations for a Hopf Bifurcation from Example 3.1 when $\beta = 0$:

$$\dot{x} = \mu x-\omega y+\alpha x(x^2+y^2)$$
and
$$\dot{y} = \omega x+ \mu y+\alpha y(x^2 + y^2)$$

Recall we can use the transformation $(x, y) = (r\cos(\theta),r\sin(\theta))$ to obtain
$$\dot{r} = \mu r + \alpha r^3$$
and
$$\dot{\theta} = \omega$$

When $\mu\alpha<0$ there is a periodic orbit in the $x-y$ plane corresponding to the fixed point $r=\sqrt{-\mu/\alpha}$. 

If we simulate the system from $x=0$ and $y = 0.8$, we get the following response.
